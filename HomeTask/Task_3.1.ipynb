```python
import pandas as pd
import numpy as np
import re
file_name = 'Energy Indicators.xls'

df_en = pd.DataFrame(pd.read_excel(file_name, usecols=range(2,6), names=['Country', 'Energy Supply', 'Energy Supply per Capita', '% Renewable']))

energy = df_en[17:244]
energy = energy.copy()
energy.iloc[:,1]=energy.iloc[:,1].where(energy.iloc[:,1]!='...')*1000000

energy.iloc[:,0] = energy.iloc[:,0].str.replace(r"[\d.]*", "", regex=True)
energy.iloc[:,0] = energy.iloc[:,0].str.replace(r"\s\(.*$", "", regex=True)
energy.replace('...', np.nan,inplace = True)

for idx,row in energy.iloc[:].iterrows():
    if row[0] == 'United States of America':
        energy.iloc[idx-17, 0] = 'United States'
    elif row[0] == 'Republic of Korea':
        energy.iloc[idx-17, 0] = 'South Korea'
    elif row[0] == 'United Kingdom of Great Britain and Northern Ireland':
        energy.iloc[idx-17, 0] = 'United Kingdom'
    elif row[0] == 'China, Hong Kong Special Administrative Region':
        energy.iloc[idx-17, 0] = 'Hong Kong'

energy.reset_index(drop=True)
```

```python
file_name_csv = 'world_bank.csv'

r = np.r_[4:60, 0]

GDP = pd.DataFrame(pd.read_csv(file_name_csv, sep=',', dtype = None, skiprows=4, usecols=r))

for idx,row in GDP.iloc[:].iterrows():
    if row[0] == 'Iran, Islamic Rep.':
        GDP.iloc[idx, 0] = 'Iran'
    elif row[0] == 'Korea, Rep.':
        GDP.iloc[idx, 0] = 'South Korea'
    elif row[0] == 'Hong Kong SAR, China':
        GDP.iloc[idx, 0] = 'Hong Kong'

GDP.rename(columns={'Country Name': 'Country'}, inplace = True)
GDP
```

```python
file_name_x = 'scimagojr.xlsx'
ScimEn = pd.DataFrame(pd.read_excel(file_name_x ))
```

```python
def answer_one():
    gdp_col = ['Country', '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015']
    ScimEn_col = ['Country', 'Rank', 'Documents', 'Citable documents', 'Citations', 'Self-citations', 'Citations per document', 'H index']

    result = pd.merge(pd.merge(GDP[gdp_col], ScimEn[ScimEn_col], how = 'outer', on = 'Country'), 
         energy, how = 'outer', on = 'Country').set_index(['Country'])
    result = result.loc[result['Rank'] < 16]
    result = result.sort_values(['Rank'])
    return result

```

```python
answer_one()
```

```python
def answer_two():
    Top15 = answer_one()
    years = []
    for i in range(2006,2015):
        years.append(str(i))
        
    avgGDP = pd.Series(Top15[years].mean(axis=1))
    avgGDP.sort_values(ascending=False, inplace=True)
    return avgGDP
```

```python
answer_two()
```

```python
def answer_three():
    col = ['2006', '2015', 0]
    res_two = pd.DataFrame(answer_two())
    t = pd.merge(answer_one(), res_two, how = 'outer', on = 'Country')
    t = t[col].sort_values(by=0, ascending=False)[0:6]
    y2006 = t['2006'].sum(axis=0)
    y2015 = t['2015'].sum(axis=0)
    res = y2015 - y2006
    return res
```

```python
answer_three()
```

```python
def answer_four():
    cit = ScimEn[['Country', 'Self-citations', 'Citations']]
    col = ['Country', 'rate']
    cit = cit.copy()
    cit['rate'] = cit['Self-citations'] / cit['Citations']
    cit = cit[col]
    cit = cit[col].sort_values(by='rate', ascending=False)[0:1]
    tup = tuple(cit.iloc[0, :])
    return tup
```

```python
answer_four()
```

```python
def answer_five():
    country = energy[['Country', 'Energy Supply', 'Energy Supply per Capita']]
    country['person'] = country['Energy Supply']/country['Energy Supply per Capita']
    country = country.sort_values(by='person', ascending=False)[2:3]
    answ = country.iloc[0, 0]
    return answ
```

```python
answer_five()
```

```python
def answer_six():
    person = energy[['Country', 'Energy Supply', 'Energy Supply per Capita']]
    person['population'] = person['Energy Supply']/person['Energy Supply per Capita']
    
    res = pd.merge(person, ScimEn, how = 'outer', on = 'Country')
    
    res['citable documents per person'] = res['Citable documents']/res['population'].where(res['population']!=0)
    
    cor = res['citable documents per person'].corr(res['Energy Supply per Capita'])
    return cor
```

```python
answer_six()
```

```python
def answer_seven():
    Top15 = answer_one()
    ContinentDict  = {'China':'Asia', 
                  'United States':'North America', 
                  'Japan':'Asia', 
                  'United Kingdom':'Europe', 
                  'Russian Federation':'Europe', 
                  'Canada':'North America', 
                  'Germany':'Europe', 
                  'India':'Asia',
                  'France':'Europe', 
                  'South Korea':'Asia', 
                  'Italy':'Europe', 
                  'Spain':'Europe', 
                  'Iran':'Asia',
                  'Australia':'Australia', 
                  'Brazil':'South America'}
    groups = pd.DataFrame(columns = ['size', 'sum', 'mean', 'std'])
    Top15['Population'] = Top15['Energy Supply'] / Top15['Energy Supply per Capita']
    for group, frame in Top15.groupby(ContinentDict):
        groups.loc[group] = [len(frame), frame['Population'].sum(),frame['Population'].mean(),frame['Population'].std()]
    return groups
```
